(3) Using k=1, report dev set accuracy for the training set sizes below. Also, measure the amount of time needed for prediction with each training size.

- time.time() gives a wall clock value you can use for timing operations

```python
def getAccuracy(preds, actual) : 
    correct, total = 0, 0
    for pred, label in zip(preds, actual):
        if pred == label: 
            correct += 1
        total += 1
    print 'total: %3d  correct: %3d  accuracy: %3.5f' %(total, correct, 1.0*correct/total)
    return 1.0*correct/total


def P3(train_sizes, accuracies):

### STUDENT START ###
    print 'k = 1'
    for s in range(len(train_sizes)):
        start = time.time()
        dSize = train_sizes[s]
        print 'size = ' + str(dSize)
        model = KNeighborsClassifier(n_neighbors=1)
        model.fit(train_data[:dSize], train_labels[:dSize])
        preds=model.predict(dev_data)
        accuracy = getAccuracy(preds, dev_labels)
        accuracies.append(accuracy)
        end = time.time()
        print 'time elapsed: '+str(end - start) + ' seconds'
    
### STUDENT END ###

train_sizes = [100, 200, 400, 800, 1600, 3200, 6400, 12800, 25000]
accuracies = []
P3(train_sizes, accuracies)
```

(6) A common image processing technique is to smooth an image by blurring. The idea is that the value of a particular pixel is estimated as the weighted combination of the original value and the values around it. Typically, the blurring is Gaussian -- that is, the weight of a pixel's influence is determined by a Gaussian function over the distance to the relevant pixel.

Implement a simplified Gaussian blur by just using the 8 neighboring pixels: the smoothed value of a pixel is a weighted combination of the original value and the 8 neighboring values. Try applying your blur filter in 3 ways:
- preprocess the training data but not the dev data
- preprocess the dev data but not the training data
- preprocess both training and dev data

Note that there are Guassian blur filters available, for example in scipy.ndimage.filters. You're welcome to experiment with those, but you are likely to get the best results with the simplified version I described above.

```python
def GBlur(inp):
    re = np.zeros(inp.shape)
    for i in range(inp.shape[0]):
        # reshapes the 784 element array into 28 x 28 matrix
        ma = np.reshape(inp[i], (28,28))
        for nrow in range(ma.shape[0]):
            for ncol in range(ma.shape[1]):
                # exlude the edges, just calculate the weighted average of the inner 27x27 pixels
                if (nrow > 0 and nrow < 27 and ncol > 0 and ncol < 27):
                    topleft = ma[nrow-1][ncol-1]
                    topcenter = ma[nrow-1][ncol]
                    topright = ma[nrow-1][ncol+1]
                    centerleft = ma[nrow][ncol-1]
                    centerright = ma[nrow][ncol+1]
                    bottomleft = ma[nrow+1][ncol-1]
                    bottomcenter = ma[nrow+1][ncol]
                    bottomright = ma[nrow+1][ncol+1]
                    ma[nrow][ncol] = 1.0/9 * (ma[nrow][ncol]+ topcenter + topleft + topright+ centerleft 
                                             +centerright + bottomleft+bottomcenter+bottomright)

        re[i] = np.reshape(ma, (784))
    return re
def P6():
    
### STUDENT START ###
    print 'P6 Gaussian Blur'
    print 'preprocess training data'
    blurred_train_data = GBlur(train_data)
    print 'preprocess dev data'
    blurred_dev_data = GBlur(dev_data)
    print 'preprocess both training and dev data'
    return GBlur(train_data)
### STUDENT END ###

P6()
```

```python

def getBlackGreyWhite(val):
    whiteThreshold = 0.33
    blackThreshold = 0.66
    if val < whiteThreshold:
        return 0
    elif (whiteThreshold<= val and val <= blackThreshold):
        return 1
    else:
        return 2

def white_grey_black_mapping(data):
    # Initialize a new feature array with the same shape as the original data.
    white_grey_black_data = np.zeros(data.shape)

    f = np.vectorize(getBlackGreyWhite)
    for feature in range(data.shape[1]):
        white_grey_black_data[:, feature] = f(data[:, feature])
    return white_grey_black_data


def P7():

### STUDENT START ###
    #blurred_train = GBlur(train_data)
    
    #BernoulliNB with binarize = 0.5
    clf = BernoulliNB(binarize=0.5)
    clf.fit(train_data, train_labels)
    preds = clf.predict(dev_data)
    # Compute accuracy on the dev data.
    print 'BernoulliNB: ', clf.score(dev_data, dev_labels)
    getAccuracy(preds, dev_labels)
    
    white_grey_black_train_data = white_grey_black_mapping(mini_train_data) 
    clf = MultinomialNB()
    clf.fit(white_grey_black_train_data, mini_train_labels)
    preds = clf.predict(dev_data)
    # Compute accuracy on the dev data.
    print 'MultinomialNB: '
    getAccuracy(preds, dev_labels)
### STUDENT END ###

P7()
```

(11) Remember that a strongly calibrated classifier is rougly 90% accurate when the posterior probability of the predicted class is 0.9. A weakly calibrated classifier is more accurate when the posterior is 90% than when it is 80%. A poorly calibrated classifier has no positive correlation between posterior and accuracy.

Train a BernoulliNB model with a reasonable alpha value. For each posterior bucket (think of a bin in a histogram), you want to estimate the classifier's accuracy. So for each prediction, find the bucket the maximum posterior belongs to and update the "correct" and "total" counters.

How would you characterize the calibration for the Naive Bayes model?

```python
def P11(buckets, correct, total):
    
### STUDENT START ###
    clf = BernoulliNB(alpha=0.0001, binarize=0.5)
    bnb = clf.fit(train_data, train_labels)
    for i in range(len(dev_data)):
        item = np.reshape(dev_data[i], (1, 784))
        pred = bnb.predict(item)
        prob = bnb.predict_proba(item)
        posterior = prob[0][int(pred)]
        for j in range(len(buckets)):
            if posterior <= buckets[j]:
                if pred == dev_labels[i]:
                    correct[j] +=1
                total[j] += 1
                
### STUDENT END ###

buckets = [0.5, 0.9, 0.999, 0.99999, 0.9999999, 0.999999999, 0.99999999999, 0.9999999999999, 1.0]
correct = [0 for i in buckets]
total = [0 for i in buckets]

P11(buckets, correct, total)

for i in range(len(buckets)):
    accuracy = 0.0
    if (total[i] > 0): accuracy = correct[i]*1.0 / total[i]
    print 'p(pred) <= %.13f    total = %3d    accuracy = %.3f' %(buckets[i], total[i], accuracy)
```

ANSWER:
In Naive Bayes, we are simply taking the posterior and update the model based on the posterior. We can calibrate our Naive Bayes classifier by changing the thresholds or the alpha values. It's mostly a trial and error.
